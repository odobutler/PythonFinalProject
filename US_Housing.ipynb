```python

#Data Imported to Colab pandas df
#source "https://www.zillow.com/research/data/"

import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
import pandas as pd
import seaborn as sns

#import housing listing prices by MSA
Housing_list= pd.read_csv("/content/drive/MyDrive/Final Project Housing Python/Metro_Median_List_Price_month.csv")


#import housing sales data by MSA
Housing_sale=pd.read_csv("/content/drive/MyDrive/Final Project Housing Python/Metro_median_Sale_price_month.csv")

```

```python
#exploring housing list and dropping first row US median
Housing_listclean =Housing_list.drop(0)
print(Housing_listclean.shape)
print(Housing_listclean)
```

```python
#exploring housing sales and dropping first row dropping first row US median
Housing_saleclean = Housing_sale.drop(0)
print(Housing_saleclean.shape)
print(Housing_saleclean.head())
```

```python

#Average Median Listing Price in 4/30/2018 - drop first row U.S. figure
Median_Listing_April_2018 = (round((Housing_listclean['4/30/2018']).mean(),))
print("Average Median Listing April 2018 =",Median_Listing_April_2018)


Median_Listing_April_2019 = (round((Housing_listclean['4/30/2019']).mean(),))
print("Average Median Listing April 2019 =",Median_Listing_April_2019)


Median_Listing_April_2020 = (round((Housing_listclean['4/30/2020']).mean(),))
print("Average Median Listing April 2020 =",Median_Listing_April_2020)


Median_Listing_April_2021 = (round((Housing_listclean['4/30/2021']).mean(),))
print("Average Median Listing April 2021 =",Median_Listing_April_2021)

Median_Listing_April_2022 = (round((Housing_listclean['4/30/2022']).mean(),))
print("Average Median Listing April 2022 =",Median_Listing_April_2022)

#Average Median Listing Price in 4/30/2020
#print(round(Housing_list['4/30/2020'].mean(),))
```

```python
#Average Median Sales Price in 4/30/2018 - drop first row U.S. figure
Median_Sales_April_2018 = (round((Housing_saleclean['4/30/2018']).mean(),))
print("Average Median Sales April 2018 =",Median_Sales_April_2018)

#Average Median Listing Price in 4/30/2019 - drop first row U.S. figure
Median_Sales_April_2019 = (round((Housing_saleclean['4/30/2019']).mean(),))
print("Average Median Sales April 2019 =",Median_Sales_April_2019)

#Average Median Listing Price in 4/30/2020 - drop first row U.S. figure
Median_Sales_April_2020 = (round((Housing_saleclean['4/30/2020']).mean(),))
print("Average Median Sales April 2020 =",Median_Sales_April_2020)

#Average Median Listing Price in 4/30/2021 - drop first row U.S. figure
Median_Sales_April_2021 = (round((Housing_saleclean['4/30/2021']).mean(),))
print("Average Median Sales April 2021 =",Median_Sales_April_2021)

#Average Median Listing Price in 4/30/2022 - drop first row U.S. figure
Median_Sales_April_2022 = (round((Housing_saleclean['4/30/2022']).mean(),))
print("Average Median Sales April 2022 =",Median_Sales_April_2022)
```

```python
#Average Median Listing Price in 4/30/2018 - drop first row U.S. figure
Median_Listing_April_2018 = (round((Housing_listclean['4/30/2018']).mode(),))
print("Mode of Median Listing April 2018 =",Median_Listing_April_2018)


Median_Listing_April_2019 = (round((Housing_listclean['4/30/2019']).mode(),))
print("Mode of Median Listing April 2019 =",Median_Listing_April_2019)


Median_Listing_April_2020 = (round((Housing_listclean['4/30/2020']).mode()))
print("Mode of Median Listing April 2020 =",Median_Listing_April_2020)


Median_Listing_April_2021 = (round((Housing_listclean['4/30/2021']).mode()))
print("Mode of Median Listing April 2021 =",Median_Listing_April_2021)

Median_Listing_April_2022 = (round((Housing_listclean['4/30/2022']).mode()))
print("Mode of Median Listing April 2022 =",Median_Listing_April_2022)
```

```python
import matplotlib.pyplot as plt
from matplotlib import numpy
from matplotlib import axis

a=Housing_listclean['4/30/2018'].plot.hist(bins=10, figsize=(10,6),alpha=1, label = "Listing 4-2018")
b=Housing_saleclean['4/30/2018'].plot.hist(bins=10, figsize=(10,6),alpha=0.6, label = "Sales 4-2018")
_=plt.title (label='US Median Listing vs Sales April 2018', fontsize=15)
_=plt.text(550000,25,'Differenece in List and Sales prices for homes were')
_=plt.ylabel('Number of MSAs')
_=plt.xlabel("Dollars")
_=plt.style.use('seaborn-white')
_=plt.xticks(fontsize=15)
_=plt.yticks(fontsize=15)
_=plt.tight_layout()
Listing = Housing_listclean['4/30/2018']
Sales= Housing_saleclean['4/30/2018']
get_label= plt.legend()
_=plt.legend(fontsize=15)

a=(round((Housing_listclean['4/30/2018']).mean(),))
b=(round((Housing_saleclean['4/30/2018']).mean(),))
d=(round((Housing_listclean['4/30/2018']).max(),))
e=(round((Housing_saleclean['4/30/2018']).max(),))


c=str(round(((a-b)/a)*100))+'%'
f=str(round(((d-e)/d)*100))+'%'

print("Average Listing",a)
print("Average Sales",b)
print(" Average Difference",c)
print ("Max Listing",d)
print ("Max Sales",e)
print (" Max Difference",f)
_=plt.text(550000,23,c) 
_=plt.annotate("Median Listing",xy=(a,20), xytext = (400000, 30), arrowprops = dict(facecolor = 'black'),color = 'b')
_=plt.annotate("Median Sales",xy=(b,20), xytext = (400000, 35), arrowprops = dict(facecolor = 'red'),color = 'r')

```
